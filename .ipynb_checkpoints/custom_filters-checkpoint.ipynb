```python
# Import resources and display image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np
from IPython import get_ipython
get_ipython().magic('matplotlib inline')
get_ipython().magic("config InlineBackend.figure_format = 'retina'")

# Read in the image
image = mpimg.imread('images/curved_lane.jpg')

plt.imshow(image)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap = 'gray')
```

```python
# Create a custom kernel
from IPython.display import Image
from IPython.display import display
# 3x3 array for edge detection
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

## TODO: Create and apply a Sobel x operator
sobel_x = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])
 
# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(gray, -1, sobel_y)
filtered_image1 = cv2.filter2D(gray, -1, sobel_x)

plt.figure()
plt.imshow(filtered_image, cmap = 'gray')
plt.figure()
plt.imshow(filtered_image1, cmap = 'gray')
```

```python
# custom filter with decimal value
image2 = mpimg.imread('images/udacity_sdc.png')
ct = np.array([[0,0,-2.5,0,0],
               [0,0,-1,0,0],
               [-2.5,-1,12,-1,-2.5],
               [0,0,-1,0,0],
               [0,0,-2.5,0,0]]) 
gray = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)
filtered_image2 = cv2.filter2D(gray, -1, ct)

plt.figure()
plt.imshow(filtered_image2, cmap = 'gray')
```
